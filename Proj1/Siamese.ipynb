```python
models_accuracy = {}
models_loss = {}
```

**Running this whole notebook again for 10 rounds for the four different settings might take some time that's why we have saved their outputs in the dicionnaries models_accuracy and models_loss . You can load them directly using pickle**

```python
with open('dictionnaries/final_models_accuracy.pickle', 'rb') as handle:
    models_accuracy = pickle.load(handle)
with open('dictionnaries/final_models_loss.pickle', 'rb') as handle:
    models_loss = pickle.load(handle)
```

```python
loss , test_acc , train_accuracy =  test_siamese_model(nb_rounds = 10 , nb_epochs = 20 , w_sharing = False , aux_loss = False , alpha = 0.1)
models_accuracy['No_Ws No_Aux'] = test_acc
models_loss['No_Ws No_Aux'] = loss
```

```python
loss , test_acc , train_acc = test_siamese_model(nb_rounds = 10 , nb_epochs = 20 , w_sharing = True , aux_loss = False , alpha = 0.1)
models_accuracy['Ws No_Aux'] = test_acc
models_loss['Ws No_Aux'] = loss
```

```python
loss , test_acc , train_acc = test_siamese_model(nb_rounds = 10 , nb_epochs = 20 , w_sharing = False , aux_loss = True , alpha = 0.1)
models_accuracy['No_Ws Aux'] = test_acc
models_loss['No_Ws Aux'] = loss
```

```python
loss , test_acc , train_acc = test_siamese_model(nb_rounds = 10 , nb_epochs = 20 , w_sharing = True , aux_loss = True , alpha = 0.1)
models_accuracy['Ws Aux'] = test_acc
models_loss['Ws Aux'] = loss
```

```python
with open('dictionnaries/final_models_accuracy.pickle', 'wb') as handle:
    pickle.dump(models_accuracy, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('dictionnaries/final_models_loss.pickle', 'wb') as handle:
    pickle.dump(models_loss, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
